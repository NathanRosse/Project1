```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Import API key
from config import chicagoCrime_key

api_url = chicagoCrime_key  # Replace with the actual API endpoint

response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract data from the API (assuming the response contains a list of records)
    data = response.json()  # You can use .text for non-JSON data

    # Create a Pandas DataFrame from the data
    df = pd.DataFrame(data)

    # Group data by "primary_type" and aggregate the results
    grouped_data = df.groupby('primary_type').size().reset_index(name='count')

    # Save the aggregated DataFrame as a CSV file
    grouped_data.to_csv('crime_data.csv', index=False)

    # Display the DataFrame
    print("Data from the API:")
    print(grouped_data)

else:
    # If the request was not successful, print an error message
    print(f"Request failed with status code {response.status_code}")
```

```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Import API key
from config import chicagoCrime_key

api_url = chicagoCrime_key  # Replace with the actual API endpoint

# Make a GET request to the API
response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract data from the API (assuming the response contains a list of records)
    data = response.json()  # You can use .text for non-JSON data

    # Create a Pandas DataFrame from the data
    df = pd.DataFrame(data)

    # Rearrange the DataFrame to have "primary_type" as the first column
    columns = list(df.columns)
    columns.remove("primary_type")
    reordered_columns = ["primary_type"] + columns
    df = df[reordered_columns]

    # Save the DataFrame as a CSV file
    df.to_csv('crime_data.csv', index=False)

    # Display the DataFrame
    print("Data from the API:")
    print(df)
else:
    # If the request was not successful, print an error message
    print(f"Request failed with status code {response.status_code}")

    
```
